```python
import pandas as pd 
```

```python
file_path = r'D:\Work\python\2000 LARGEST CAMPANIES\largest campanies 2022.xlsx'
df = pd.read_excel(file_path)
```

```python
df
```

```python
def convert_to_billions(sales):
    # Strip any leading or trailing whitespace
    sales = sales.strip()
    sales = sales.replace(',', '')

    
    if 'B' in sales:
        return float(sales.replace('$', '').replace(' B', '').strip())
    elif 'M' in sales:
        return float(sales.replace('$', '').replace('M', '').strip()) / 1000
    else:
        raise ValueError("Unknown format")

# Apply the conversion function to the 'sales' column
df['assets_in_billions'] = df['Assets'].apply(convert_to_billions)
df['marketValue_in_billions']=df['Market Value'].apply(convert_to_billions)
df['profit_in_billions']= df['profit'].apply(convert_to_billions)
df['sales_in_billions']= df['Sales'].apply(convert_to_billions)



df
```

```python
import os 
#urpose: This function retrieves the directory name from the full file path specified in file_path.
#Example: If file_path is r'D:\Work\python\2000 LARGEST CAMPANIES\largest campanies 2022.xlsx', then os.path.dirname(file_path) will return r'D:\Work\python\2000 LARGEST CAMPANIES'.
output_file_name = 'largest_companies_2022_processed.xlsx'  # Name of the new file
output_path = os.path.join(os.path.dirname(file_path), output_file_name)

# Save the modified DataFrame to a new Excel file
try:
    df.to_excel(output_path, index=False)  # Set index=False to avoid writing row indices
    print(f"DataFrame saved to {output_path}")
except PermissionError as e:
    print(f"Permission error: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```
